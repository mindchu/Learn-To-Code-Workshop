### Fun Fact
Python เป็นภาษาที่ Case-sensitive หมายความว่าชื่อเดียวกันแต่พิมพ์ต่างกัน คือคนละตัวแปร

### **Arithmetic Operator**
เราสามารถนำ **ตัวแปร (variables)** หรือ **ค่าข้อมูล (values)** มาคำนวณทางคณิตศาสตร์ได้ด้วย **Arithmetic Operators** โดยแบ่งเป็นดังนี้:
| เครื่องหมาย | ความหมาย | ตัวอย่าง |
|-------------|-----------|-----------|
| `+` | บวก (Addition) | `a + b` |
| `-` | ลบ (Subtraction) | `a - b` |
| `*` | คูณ (Multiplication) | `a * b` |
| `/` | หาร (Division) | `a / b` |
| `//` | หารปัดเศษลง (Floor Division) | `a // b` |
| `%` | หารเอาเศษ (Modulus) | `a % b` |
| `**` | ยกกำลัง (Exponentiation) | `a ** b` |

````python
a = 17
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** 2 = {a ** 2}")
````

### **Comparison Operator**
เราสามารถนำ **ตัวแปร (variables)** หรือ **ค่าข้อมูล (values)** มาเปรียบเทียบกันด้วย **Comparison Operators** โดยผลลัพธ์ที่ได้จะเป็น `True` (จริง) หรือ `False` (เท็จ)

เครื่องหมายที่ใช้เปรียบเทียบมีดังนี้:
| เครื่องหมาย | ความหมาย | ตัวอย่าง |
|-------------|-----------|-----------|
| `==` | เท่ากับ | `a == b` |
| `!=` | ไม่เท่ากับ | `a != b` |
| `>` | มากกว่า | `a > b` |
| `<` | น้อยกว่า | `a < b` |
| `>=` | มากกว่าหรือเท่ากับ | `a >= b` |
| `<=` | น้อยกว่าหรือเท่ากับ | `a <= b` |

````python
a = 17
b = 5

print(f"{a} == {b} -> {a == b}")
print(f"{a} != {b} -> {a != b}")
print(f"{a} > {b} -> {a > b}")
print(f"{a} < {b} -> {a < b}")
print(f"{a} >= {b} -> {a >= b}")
print(f"{a} <= {b} -> {a <= b}")
````

## บทที่ 3 • ข้อความ (String)

### **Formatted String (f-string)**

ในบางครั้งที่เราอยากรวม **ข้อความ (text)** and **ตัวแปร (variables)** อยู่ในบรรทัดเดียวกัน

Python มีสิ่งที่เรียกว่า **f-string** ซื่งทำให้เราสามารถนำค่าจากตัวแปรมาแทรกในข้อความได้ โดยใช้ `{}`.

**Syntax**

```python
f"... {variable} ..."

````python
# Without f-string
name = "Mind"

print("Hello " + name)
````

````python
# With f-string. More easy to read!
name = "Mind"

print(f"Hello {name}")
````

````python
name = "Mind"
age = 20
major = "Computer Engineering"

print(f"My name is {name}. I am {age} years old and I study {major}.")
````

````python
# calculate inside {}
price = 150
qty = 3

print(f"Total: {price} x {qty} = {price * qty} THB")
````

### Tip
ยังมีวิธีให้เราเล่นกับ String ได้อีกมากมาย น้อง ๆ สามารถลองไปศึกษานอกรอบดูได้นะ

````python
text = "Python"
print(text.upper())
print(text.lower())
print(len(text))
print(text[0])       # first char
print(text[-1])      # last char
print(text[0:3])     # slice 0 to before 3
print(text[::-1])    # reverse
print("Py" in text)  # contains "Py"?
````

## บทที่ 4 • เงื่อนไข (if / elif / else)

### **if, elif และ else**

คำสั่ง **if** ใช้สำหรับตรวจสอบเงื่อนไข และตัดสินใจว่าจะให้โปรแกรมทำอะไร

เงื่อนไขจะต้องมีผลลัพธ์เป็น `True` หรือ `False`

- `if` ใช้ตรวจสอบเงื่อนไขแรก
- `elif` ใช้ตรวจสอบเงื่อนไขถัดไป หากเงื่อนไขก่อนหน้าเป็น `False`
- `else` ใช้เมื่อไม่มีเงื่อนไขใดเป็น `True`

**รูปแบบการเขียน**

```python
if เงื่อนไข:
    คำสั่ง

elif เงื่อนไข:
    คำสั่ง

else:
    คำสั่ง
```

> **หมายเหตุ:** Python ใช้ **Indentation (การเว้นวรรค)** เพื่อบอกว่าคำสั่งใดอยู่ภายใน `if`, `elif` หรือ `else`

````python
age = 20

if age >= 18:
    print(f"Your age is {age}. You're an adult.")
else:
    print(f"Your age is {age}. You're under 18.")
````

````python
weather = "rainy"

if weather == "sunny":
    print(f"The weather is {weather}. Go outside.")
elif weather == "rainy":
    print(f"The weather is {weather}. Bring an umbrella.")
else:
    print(f"The weather is {weather}. Stay at home.")
````

````python
score = 78

if score >= 80:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
elif score >= 50:
    grade = "D"
else:
    grade = "F"
print(f"{score} → {grade}")
````
